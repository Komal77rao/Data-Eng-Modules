# Customers who never order

### Loading the data

````python
customers = {"headers": ["id", "name"], "values": [[1, "Joe"], [2, "Henry"], [3, "Sam"], [4, "Max"]]}
````

````python
orders = {"headers": ["id", "customerId"], "values": [[1, 3], [2, 1]]}
````

````python
from time import process_time
start_time = process_time()
import subprocess
try:
    import google.colab # if colab exists, install pysqlite-binary
    subprocess.run(['pip', 'install', 'pysqlite3-binary'], capture_output=False)
    import pysqlite3 as sqlite3

except ModuleNotFoundError:
    pass
end_time = process_time()

import warnings
warnings.simplefilter(action='ignore', category=UserWarning)
import pandas as pd
````

````python
import sqlite3

conn = sqlite3.connect('crm.db')

import pandas as pd
def build_dataframe(data):
    columns = data['headers']
    values = data['values']
    df = pd.DataFrame(values)
    df.columns = columns
    return df
````

````python
customers_df = build_dataframe(customers)
customers_df
````

````python
orders_df = build_dataframe(orders)
orders_df
````

````python
customers_df.to_sql('customers', conn, if_exists = 'replace', index = False)
````

````python
orders_df.to_sql('orders', conn, if_exists = 'replace', index = False)
````

````python
pd.read_sql('select * from customers', conn)
````

````python
pd.read_sql('select * from orders', conn)
````

````python
query = """
select * from customers c
where id not in (select customerId from orders)
"""

pd.read_sql(query, conn)
````
